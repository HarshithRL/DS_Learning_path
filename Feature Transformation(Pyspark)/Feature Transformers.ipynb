What is Vectorization in Machine Learning?

Vectorization is jargon for a classic approach of converting input data from its raw format (i.e. text ) into vectors of real numbers which is the format that ML models support.

In spark that supports Feature Transformers are VectorAssembler ,OneHotEncoder , Imputer and more ......

Ex:- Lets Show Example to Vectorize the Numerical Columns 

```
#Assembling all the numeric column into one vector
from pyspark.ml.feature import VectorAssembler

vec_assembler = VectorAssembler(inputCols=["AGE",'SMOKING','YELLOW_FINGERS','ANXIETY','PEER_PRESSURE','CHRONIC DISEASE','FATIGUE ','ALLERGY '], outputCol="features")

vec_train_df = vec_assembler.transform(train_df)
```
